```
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.RandomState(seed=42)
pi=3.14

u1= rng.uniform(0,1,size=10000)
u2=rng.uniform(0,1,size=10000)
s = -np.log(u1)
theta = 2*pi*u2
r = np.sqrt(2*s)
x = r*np.cos(theta)
y= r*np.sin(theta)
```

```
plt.scatter(x, y,color='green')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(x,bins=60,color='blue')
plt.title("BOX MULLER SAMPLE X")
plt.xlabel("x")
plt.ylabel("frequency")
```

```
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(y,bins=60,color='red')
plt.title("BOX MULLER SAMPLE Y")
plt.xlabel("y")
plt.ylabel("frequency")
```
